## Логирование

**Logging** или **логирование** (журналирование) - это автоматическое ведение записей о тех или иных событиях в хронологическом порядке с целью либо убедиться в том, что все идет по плану, либо упростить отлавливание нежелательного поведения программного обеспечения. Когда что-то идет не так, опытные разработчики смотрят логи, чтобы понять где, когда и при каких условиях возникла та или иная ситуация.

Логи удобно использовать не только на этапе разработки и отладки приложения, но и в продакшне, где нежелательно останавливать работу и запускать какой-нибудь дебаггер по каждому случаю, требующему внимания программиста и/или системного администратора.

Альтернативный способ мониторинга состояния проекта - использование "принтов". Однако у этого способа немало недостатков, среди которых:
- Нельзя настроить единообразное форматирование вывода логов. Точнее, это не получится сделать удобно в одном месте, применив ко всему проекту единоразово
- Нельзя легко управлять уровнями логирования.
- Неудобно записывать историю состояния в файл.
- И многое другое.

#### Поговорим о стандартной библиотеке logging
У данной библиотеки довольно хорошая документация, в которую вы всегда можете углубиться для лучшего понимания настройки и работы с инструментом.

А самая простая программа с использованием библиотеки может выглядеть так:

```python
import logging

logging.error('Это лог какой-то ошибки')
```

В данном сообщении `ERROR` - это уровень, на котором был создан лог, `root` - это имя логгера, а так как мы еще не создавали свои логгеры, то в нашем случае, это `root`, то есть корневой логгер, ну, и сам текст лога, который мы передали - "Это лог какой-то ошибки".

Основными ключевыми понятиями в библиотеке `logging` являются:

**Логгеры**. Предоставляют интерфейс пользователю для создания логов в коде.  
**Уровни логирования**. Определяют степень серьезности информации, выводимой в логах.  
**Хэндлеры**. Определяют место вывода логов (например, консоль или файл).  
**Форматтеры**. Определяют то, как логи будут выглядеть.  
**Фильтры**. Позволяют гибко настраивать то, какие логи выводить.
